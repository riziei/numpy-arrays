```python
import numpy as np
```

creating arrays 1d and 2d

```python
# 1D array
arr1 = np.array([10, 20, 30, 40])
print(arr1)

# 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)
```

array properties<br>
shape (rows and columns), number of dimensions, total elements, and data type

```python
print(arr2.shape)   # rows, columns
print(arr2.ndim)    # dimensions
print(arr2.size)    # total elements
print(arr2.dtype)   # data type
```

array indexing

```python
print(arr1[0])      # first element
print(arr2[1][2])   # row 1, col 2
```

array slicing <br>
Slicing allows us to extract a portion of an array using a range of indices


```python
print(arr1[1:3])
print(arr2[:, 1])   # second column
```

mathematical operations on array

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("addition : ",a + b)
print("subtraction : ",a - b)
print("multiplication : ",a * b)
print("division : ",a / b)
```

scalar operations 

```python
print('original arry a: ',a)
print('original arry b: ',b)
print("adding scalar 10 a: ",a + 10)
print('multiplying scalar 2 b: ',a * 2)
```

array functions <br>
built-in functions like sum, mean, max, min, and standard deviation

```python
arr = np.array([10, 5, 8, 20, 3])

print('sum :',np.sum(arr))
print('mean :',np.mean(arr))
print('maximum :',np.max(arr))
print('minimun :',np.min(arr))
print('standare deviation :',np.std(arr))
```

reshaping arrays

```python
arr = np.array([1, 2, 3, 4, 5, 6])
print("array before reshaping:",arr)
reshaped = arr.reshape(2, 3) #convert arrays into 2 rows and 3colunms
print("new reshaped array:",reshaped)
```

random arrays<br>
random arras are generated using np.random.randint(low, high=None, size=None, dtype=int)

```python
random_arr = np.random.randint(1, 100, size=10) # 1 to 100 ,100 excluded
print("random array : ",random_arr)
```

filtering array

```python
arr = np.array([10, 25, 30, 45, 50])
print("the array: ",arr)
print('elements greater than 30:',arr[arr > 30])
```
